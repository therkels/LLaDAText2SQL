```python
import os
import importlib
from database_interface import DatabaseHelper
from get_raw_data import get_raw_data, convert_data_to_namedtuples
from dotenv import load_dotenv
from tqdm.notebook import tqdm, trange
load_dotenv()
```

# 1. Configure the DB once so that we have the data loaded


Get the databse stuff setup

```python
db_dir = os.getenv("DB_DIR")
print(f"db directory: {db_dir}")
if not db_dir:
    raise ValueError("DB_DIR environment variable not set.")
db_name = os.getenv("DB_NAME")
print(f"db name: {db_name}")
if not db_name:
    raise ValueError("DB_NAME environment variable not set.")

db_helper = DatabaseHelper(os.path.join(db_dir, db_name))
print("db helper setup complete")
```

# 2. Get Dataset we are loading

```python
file_path = os.getenv("RAW_DATA_PATH")
if not file_path:
    raise ValueError("RAW_DATA_PATH environment variable not set.")
data = get_raw_data(file_path)
for idx, data_row in enumerate(tqdm(convert_data_to_namedtuples(data), total=len(data))):
    # print(f"Row {idx}: {data_row}")
    db_helper.insert_data(data_row.sql_context)


```

# 3. Do we have data?

```python
print(f"number of tables in db instance: {db_helper.get_table_count()}")

select_query = 'SELECT * from satellites;'
results = db_helper.fetch_data(select_query)
print(results)
```
